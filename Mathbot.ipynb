```python
#maths with quadratic and cubic
import sympy as sp
import random
import matplotlib.pyplot as plt
import numpy as np
from PIL import ImageTk, Image
from sympy import symbols
import threading

x = symbols('x')

responses = [
            "The answer is {result}",
            "It's {result}",
            "The result is {result}",
            "You get {result} as the solution",
            "The solution is {result}",
            "After performing the operation, you get {result}",
            "The calculated value is {result}",
            "You obtain {result} as the answer",
            "{result} is the outcome of the calculation",
            "By doing the calculation, we arrive at {result}",
            "The final result is {result}",
            "After evaluating the expression, the answer is {result}"
        ]

def solve_cubic_equation(equation):
    try:
        x = symbols('x')
        equation_parts = equation.split('=')
        lhs = sp.sympify(equation_parts[0].strip().replace('^', '**'))
        rhs = sp.sympify(equation_parts[1].strip().replace('^', '**'))
        cubic_eq = sp.Eq(lhs, rhs)
        result = sp.solve(cubic_eq, x)
        result_str = str(result).replace("**", "^").replace("sqrt","√")
        return random.choice(responses).format(result=result_str)
    except (sp.SympifyError, ValueError):
        return "Invalid cubic equation"

def solve_equation(equation):
    try:
        x = symbols('x')
        equation_parts = equation.split('=')
        lhs = sp.sympify(equation_parts[0].strip())
        rhs = sp.sympify(equation_parts[1].strip())
        equation = sp.Eq(lhs, rhs)
        result = sp.solve(equation, x)
        result_str = str(result).replace("**", "^").replace("sqrt","√")
        return random.choice(responses).format(result=result_str)
    except (sp.SympifyError, ValueError):
        return "Invalid equation"

def solve_quadratic_equation(equation):
    try:
        x = symbols('x')
        equation_parts = equation.split('=')
        lhs = sp.sympify(equation_parts[0].strip().replace('^', '**'))
        rhs = sp.sympify(equation_parts[1].strip().replace('^', '**'))
        quadratic_eq = sp.Eq(lhs, rhs)
        result = sp.solve(quadratic_eq, x)
        result_str = str(result).replace("**", "^").replace("sqrt","√")
        return random.choice(responses).format(result=result_str)
    except (sp.SympifyError, ValueError):
        return "Invalid quadratic equation"

def solve_derivative(expression):
    try:
        expression_str = str(expression)
        expression_str = expression_str.replace("^", "**")
        result = sp.diff(expression_str, x)
        result_str = str(result).replace("**", "^").replace("sqrt","√")
        return random.choice(responses).format(result=result_str)
    except sp.SympifyError:
        return "Invalid expression"

def solve_integral(expression):
    try:
        expression = expression.replace("^", "**")
        expression = re.sub(r'(\d+)x', r'\1*x', expression)
        integral = sp.integrate(expression, x)
        integral_solution = str(integral).replace("**", "^")+ " + C"
        return random.choice(responses).format(result=integral_solution)
    except (sp.SympifyError, ValueError):
        return "Invalid expression"


def plot_quadratic_equation(equation, image_name):
    x = np.linspace(-10, 10, 400)
    
    equation_parts = equation.split('=')
    lhs = equation_parts[0].strip().replace('^', '**')
    rhs = equation_parts[1].strip().replace('^', '**')

    y = eval(lhs) - eval(rhs)

    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Quadratic Equation')
    plt.grid(True)
    
    # Save the graph as an image
    image_path = f"{image_name}.png"
    plt.savefig(image_path)
    
    # Close the plot to prevent it from being displayed in the terminal
    plt.close()
    
    # Return the image path
    return image_path

def plot_cubic_equation(equation, image_name):
    x = np.linspace(-10, 10, 400)
    equation_parts = equation.split('=')
    lhs = equation_parts[0].strip().replace('^', '**')
    rhs = equation_parts[1].strip().replace('^', '**')

    y = eval(lhs) - eval(rhs)

    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Cubic Equation')
    plt.grid(True)
    
    # Save the graph as an image
    image_path = f"{image_name}.png"
    plt.savefig(image_path)
    
    # Close the plot to prevent it from being displayed in the terminal
    plt.close()
    
    # Return the image path
    return image_path


def process_user_input(event=None):
    user_input = user_entry.get()
    user_entry.delete(0, tk.END)
    # Check if the user input is a direct equation or a request to solve an equation
    if "=" in user_input:
        if "x^2" in user_input:
            solution = solve_quadratic_equation(user_input)
            # Generate and display the graph
            image_name = "quadratic_graph"
            image_path = plot_quadratic_equation(user_input, image_name)
        
        elif "x^3" in user_input:
            solution = solve_cubic_equation(user_input)
             # Generate and display the graph
            image_name = "cubic_graph"
            image_path = plot_cubic_equation(user_input, image_name)
        else:
            solution = solve_equation(user_input)
            image_path = 0
            
        # Convert the solution to a readable format
        if isinstance(solution, tuple):
            solution_x = ', '.join(str(sol) for sol in solution[0])
            response = f"Solutions for x: {solution_x}"
        else:
            response = solution
     # Display the image in the chat history
        if image_path:
            image = Image.open(image_path)
            image.thumbnail((520, 320))  
            
            # Convert the image to Tkinter-compatible format
            image_tk = ImageTk.PhotoImage(image)
            
            # Insert the image in the chat history
            chat_text.image_create(tk.END, image=image_tk)
            chat_text.image = image_tk
        
    elif "dy/dx" in user_input.lower() or "derivative of" in user_input.lower():
        expression = user_input.lower().replace("dy/dx", "").replace("derivative of", "")
        derivative_solution = solve_derivative(sp.sympify(expression))
        response = str(derivative_solution)
    
    elif "integral of " in user_input.lower():
        expression = user_input.lower().replace("integral of", "").strip()
        integral_solution = solve_integral(expression)
        response = integral_solution

    elif re.search(r'(\d+)\s*([\+\-\*/\^%])\s*(\d+)', user_input):
        match = re.search(r'(\d+)\s*([\+\-\*/\^%])\s*(\d+)', user_input)
        num1 = float(match.group(1))
        operator = match.group(2)
        num2 = float(match.group(3))

        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            result = num1 / num2
        elif operator == '^':
            result = num1 ** num2
        elif operator == '%':
            result = num1 % num2
            
        response = random.choice(responses).format(result=result)

    else:
        # Preprocess the user input
        preprocessed_query = preprocess_text(user_input)
        flattened_query = ' '.join(preprocessed_query)

        # Vectorize the query
        query_vector = vectorizer.transform([flattened_query])

        # Calculate the cosine similarities between the query and dataset vectors
        similarities = cosine_similarity(query_vector, vectorizer.transform(flattened_dataset))

        # Find the index of the most similar question in the dataset
        most_similar_index = similarities.argmax()

        # Get the corresponding answers
        answers = dataset[most_similar_index]['answers']
        response = random.choice(answers)  

    # Display user input and chatbot response in the chat history
    chat_text.tag_config("user", foreground="#0078FF", justify="right")
    chat_text.tag_config("chatbot", foreground="#555555", justify="left")

    chat_text.insert(tk.END, user_input + "\n", "user")

    if isinstance(response, list):
        response_str = ' '.join(map(str, response))
        chat_text.insert(tk.END, response_str + "\n\n", "chatbot")
    else:
        chat_text.insert(tk.END, response + "\n\n", "chatbot")
        
    # Scroll to the end of the chat history
    chat_text.see(tk.END)
        
    # Update the chat history before converting the response to speech
    chat_text.update_idletasks()
    
    
    response_for_speech = response.replace("√", "square root of").replace("-", "minus").replace("^","raised to").replace("*","multiplied by").replace("I/","I divided by").replace("x/","x divided by")
    
    def speak_response():
        engine.say(response_for_speech)
        engine.runAndWait()
    
    # Start a new thread to run the speech engine
    speech_thread = threading.Thread(target=speak_response)
    speech_thread.start()
    
```
